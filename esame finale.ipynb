```python
import pandas as pd
#il caricamento del dataset
file_path = r'C:\Users\moujd\Downloads\owid-covid-data.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
#dopo che ho letto bene l esame prima di iniziare ,che c e da fare attenzione a qualche colonne 
print(df['continent'].unique())

# verifico la colonna location per vedere le regioni e pesi che sono presenti nel dataset
print(df['location'].unique())

# Verifico la presenza di valori nulli nelle colonne 'new_cases' e 'total_cases'
print(df[['new_cases', 'total_cases']].isnull().sum())

# Verifico se ci sono valori anomali in total_cases
print(df[df['total_cases'] < 0])
```

```python
#qua pulisco cio che non serve o non e nel suo colonna guista
locations_da_cancellare = ['Africa', 'Asia', 'World', 'Oceania', 'North America', 'South America', 'Europe']
df_da_pulire = df[~df['location'].isin(locations_da_cancellare)]

# Converto la colonna 'date' in formato datetime
df_da_pulire.loc[:, 'date'] = pd.to_datetime(df_da_pulire['date'])

# Rimuovo le righe con valori nulli nelle colonne new_cases e total_cases
df_da_pulire = df_da_pulire.dropna(subset=['new_cases', 'total_cases'])
```

```python
#adesso inizio a rispondere alle domande dopo che ho pulito il mio dataset
#il numero di casi fin dall'inizio della pandemia
casi_per_continente = df.groupby('continent')['total_cases'].sum()
casi_per_continente
```

```python
#la percentuale rispetto al totale mondiale del numero di casi
tot = df['total_cases'].sum()
print(f"il Totale mondiale dei casi,{ tot}")

percentuale_casi_continente = (casi_per_continente / tot) * 100
print(percentuale_casi_continente)
```

```python
#Selezionare i dati relativi all'Italia nel 2022 e, poiché i nuovi casi vengono registrati settimanalmente.
df['date'] = pd.to_datetime(df['date'])
italia = df.loc[(df['location'] == 'Italy') & (df['date'].dt.year == 2022)]
italia.head()
```

```python
# Filtro i giorni con dati mancanti per 'new_cases'
italia_2022 = italia.dropna(subset=['new_cases'])
italia_2022.head(3)
```

```python
# Creo il grafico per l'evoluzione del casi totali dall'inizio alla fine dell'anno
import seaborn as sns
sns.set(style="whitegrid")
sns.lineplot(x='date', y='total_cases', data=italia_2022)
```

```python
#l numero di nuovi casi rispetto alla data
```

```python
# Creo il grafico dei nuovi casi giornalieri utilizzando Seaborn
g = sns.lineplot(x='date', y='new_cases', data=italia_2022, marker='o', color='red')
g.set(title='Nuovi Casi Giornalieri in Italia nel 2022', xlabel='Data', ylabel='Nuovi Casi')

sns.despine()
```

```python
# un boxplot in cui vediamo la differenza tra nazioni che sono state chiesti

df['date'] = pd.to_datetime(df['date'])

countries = ['Italy', 'Germany', 'France']
filtro = df[(df['location'].isin(countries)) & 
                 (df['date'] >= '2022-05-01') & 
                 (df['date'] <= '2023-04-30')]
sns.boxplot(x='location', y='icu_patients', data=filtro, palette='Set2', hue='location', legend=False)
```

```python
#un breve commento riguarda cio che ho osservato nel grafico
#ho osservato che la Germania ha un numero maggiore di pazienti ICU  rispetto a Francia e Italia
```

```python
#domanda:mostrare, in maniera grafica oppure numerica, la somma dei pazienti ospitalizzati per ognuna (colonna hosp_patients)
df_da_pulire['date'] = pd.to_datetime(df_da_pulire['date'], errors='coerce')

df_2021 = df_da_pulire[(df_da_pulire['date'].dt.year == 2021) & 
                     (df_da_pulire['location'].isin(['Italy', 'Germany', 'France', 'Spain']))]

pazienti_ospitalizzati = df_2021.groupby('location')['hosp_patients'].sum()

# per la risposta numerica 
print(pazienti_ospitalizzati)
#questa in maniera grafica
sns.barplot(x=pazienti_ospitalizzati.index, y=pazienti_ospitalizzati.values)
sns
```
